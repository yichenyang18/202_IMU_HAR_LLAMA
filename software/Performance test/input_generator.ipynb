```python
directory_path = r'D:\course_grad\202_embedded\dataset\HAR\UCI HAR Dataset\train\Inertial Signals'

train_data = {}
train_names = []
interval = 4
num_samples = 20

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory_path, filename)
        file_lines = []
        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip().split()
                line = [line[i * interval] for i in range(num_samples)]
                file_lines.append(list(map(lambda x: round(float(x), 4), line)))
        train_data[filename[:-4]] = file_lines
        train_names.append(filename[:-4])
```

```python
time_duration = 0.02 * interval * (num_samples - 1)
time_duration
```

```python
def calc_mean_variance(values):
    return round(sum(values) / len(values), 4), round(pd.Series(values).var(), 4)
```

```python
filename = 'llama_inputs/inputs_enriched_num_samples_{}.txt'.format(num_samples)
with open(filename, 'w') as f:
    for index, row in df.iterrows():
        if index in rows:
            x_acc = row['body_acc_x_train']
            y_acc = row['body_acc_y_train']
            z_acc = row['body_acc_z_train']
            x_gyro = row['body_gyro_x_train']
            y_gyro = row['body_gyro_y_train']
            z_gyro = row['body_gyro_z_train']

            xacc_mean, xacc_var = calc_mean_variance(x_acc)
            yacc_mean, yacc_var = calc_mean_variance(y_acc)
            zacc_mean, zacc_var = calc_mean_variance(z_acc)
            xgyro_mean, xgyro_var = calc_mean_variance(x_gyro)
            ygyro_mean, ygyro_var = calc_mean_variance(y_gyro)
            zgyro_mean, zgyro_var = calc_mean_variance(z_gyro)    

            x_acc_str = ', '.join([f"{x:.3f} g" for x in x_acc])
            y_acc_str = ', '.join([f"{y:.3f} g" for y in y_acc])
            z_acc_str = ', '.join([f"{z:.3f} g" for z in z_acc])
            x_gyro_str = ', '.join([f"{x:.3f} g" for x in x_gyro])
            y_gyro_str = ', '.join([f"{y:.3f} g" for y in y_gyro])
            z_gyro_str = ', '.join([f"{z:.3f} g" for z in z_gyro])  

            prompt = f"IMU data: 1.Triaxial acceleration signal: X-axis: {x_acc_str}; Y-axis: {y_acc_str}; Z-axis: {z_acc_str}; X-axis-mean={xacc_mean} g, X-axis-var={xacc_var}; Y-axis-mean={yacc_mean} g, Y-axis-var={yacc_var}; Z-axis-mean={zacc_mean} g, Z-axis-var={zacc_var}"
            prompt2 = f" 1.Triaxial angular velocity signal: X-axis: {x_gyro_str}; Y-axis: {y_gyro_str}; Z-axis: {z_gyro_str}; X-axis-mean={xgyro_mean} rad/s, X-axis-var={xgyro_var}; Y-axis-mean={ygyro_mean} rad/s, Y-axis-var={ygyro_var}; Z-axis-mean={zgyro_mean} rad/s, Z-axis-var={zgyro_var}"
            prompt3 = f" 1.Triaxial acceleration signal: The provided three-axis acceleration signals contain acceleration data for the X-axis, Y-axis, and Z-axis respectively. Each axis's data is a time-series signal consisting of some data samples, measured at a fixed time interval with a frequency of 10Hz(10 samples is collected per second) by smartphone (Samsung Galaxy S II) wearing on the waist. The unit is gravitational acceleration (g), equivalent to 9.8m/s^2. It's important to note that the measured acceleration is influenced by gravity, meaning the acceleration measurement along a certain axis will be affected by the vertical downward force of gravity. 2. Triaxial angular velocity signal: The provided three-axis angular velocity signals contain angular velocity data for the X-axis, Y-axis, and Z-axis respectively. Each axis's data is a time-series signal consisting of some data samples, measured at a fixed time interval with a frequency of 10Hz. The unit is radians per second (rad/s)."
            prompt4 = f" Please analyze what activity the person wearing this IMU is doing. (options: WALKING, SITTING). Return your analysis and add your answer as \"ANSWER: (ONLY ONE OF THE OPTIONS THAT YOU CHOOSE)\" at the end (no need for any coding).\n"
            
            final_prompt = prompt + prompt2 + prompt3 + prompt4
            f.write(final_prompt)
```

```python
directory_path = r'D:\course_grad\202_embedded\dataset\HAR\UCI HAR Dataset\train\Inertial Signals Denoised'

train_data = {}
train_names = []
interval = 20
num_samples = 6

for filename in os.listdir(directory_path):
    if filename.endswith('.txt'):
        file_path = os.path.join(directory_path, filename)
        file_lines = []
        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip().split()
                tsp, sb = map(lambda x: round(float(x), 4), line[:2])
                line = line[2:]
                line = [line[i * interval] for i in range(num_samples)]
                lst = [tsp, sb]
                lst.extend(list(map(lambda x: round(float(x), 4), line)))
                file_lines.append(lst)
        train_data[filename[:-4]] = file_lines
        train_names.append(filename[:-4])
```

```python
time_duration = 0.02 * interval * (num_samples - 1)
time_duration
```
